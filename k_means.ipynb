```python
# ----------- K-means Clustering for Similarity Recognition -------------

import numpy as np
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load dataset
iris = load_iris()
X = iris.data

print("Dataset shape:", X.shape)

# 2. Scaling (important for clustering)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Apply K-Means clustering
k = 3  # number of clusters
kmeans = KMeans(n_clusters=k, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

print("\nCluster labels assigned by K-Means:")
print(clusters)

# 4. Evaluate clustering performance
silhouette = silhouette_score(X_scaled, clusters)
print("\nSilhouette Score (higher = better clustering):", silhouette)

# 5. Visualizing Clusters (first two features)
plt.figure(figsize=(7,5))
sns.scatterplot(
    x=X_scaled[:, 0],
    y=X_scaled[:, 1],
    hue=clusters,
    palette="Set1",
    s=80
)
plt.title("K-Means Clustering (Similarity Recognition)")
plt.xlabel("Feature 1 (scaled)")
plt.ylabel("Feature 2 (scaled)")
plt.legend(title="Cluster")
plt.show()

# 6. Show cluster centers
print("\nCluster Centers (in scaled feature space):")
print(kmeans.cluster_centers_)

# 7. Example: Identify similarity
sample = X_scaled[0].reshape(1, -1)
cluster_of_sample = kmeans.predict(sample)

print("\nExample:")
print("The first data point belongs to Cluster:", cluster_of_sample[0])
```
